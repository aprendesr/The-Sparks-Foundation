#### 1. Import the required libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 
from sklearn import metrics  
print ("The required libraries were successfully imported")
```

#### 2. Access the data through the link provided for the task: http://bit.ly/w-data

```python
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
print(data.shape)
print(data.info())
print(data.head())
print("The required data was sussessfully imported")
```

#### 3. Visualize the imported data in a scatterplot

```python
data.plot(x='Hours',y='Scores', style='o', c='b')
plt.title('Relationship between hours studied per day and achieved score in percentage')
plt.xlabel('Hours studied per day')
plt.ylabel('Achieved score in percentage')
```

#### 5. Perform simple linear regression

```python
X = data.iloc[:, :-1].values  
Y = data.iloc[:, 1].values
print("Data was successfully preprocessed")
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
print("Data set was successfully divided into two sub-sets: train and test")
```

```python
linearregression = LinearRegression()  
linearregression.fit(X_train, Y_train) 
print("The model training was successfully completed")
```

```python
regressionline = linearregression.coef_ * X + linearregression.intercept_
plt.scatter(x=X, y=Y, c='b')
plt.plot(X,regressionline, c='r')
plt.title('Regression line between hours studied per day and achieved score in percentage')
plt.xlabel('Hours studied per day')
plt.ylabel('Achieved score in percentage')
plt.show
```

#### 6. Make predictions with the build model

```python
print(X_test)
Y_pred = linearregression.predict(X_test)
```

```python
df = pd.DataFrame({'Actual scores': Y_test, 'Predicted scores': Y_pred})  
df.head()
```

```python
df.plot(kind='bar',figsize=(6,6), color=['b', 'r'])
plt.show()
```

#### 7. Introduce a number of studied hours per day to obtain the predicted score

```python
while True:
    Hours = input("Please enter the studied hours per day to estimate the score: ")
    try:
        Hours = int(Hours)
        Score=linearregression.coef_*(Hours)+linearregression.intercept_
        if Hours <0:
            print("Please enter a valid number of hours")
        elif Score >100:
            print ("The predicted score of a student that spent ",Hours,"hours studying per day is: 100 ")
        else:
            print ("The predicted score of a student that spent ",Hours,"hours studying per day is:", Score)
        break;
    except ValueError:
        try:
            Hours =float(Hours)
            Score=linearregression.coef_*(Hours)+linearregression.intercept_
            if Hours <0:
                print("Please enter a valid number of hours")
            elif Score >100:
                print ("The predicted score of a student that spent ",Hours,"hours studying per day is: 100 ")
            else:
                print ("The predicted score of a student that spent ",Hours,"hours studying per day is:", Score)
            break;
        except ValueError:
            print("Please enter a valid number number of hours")
```

#### 8. Evaluate the model

```python
print('Mean Absolute Error:',metrics.mean_absolute_error(Y_test, Y_pred)) 
print('Mean Squared Error:', metrics.mean_squared_error(Y_test, Y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Y_test, Y_pred)))
print('R-2:', metrics.r2_score(Y_test, Y_pred))
```

Conclusion: the task requested to make a prediction by using Supervised ML. In order to do this, it was necesary to import the data set and visualize it. There were only two variables involved: the number of studied hours per day and the achieved score and the scatterplot showed a positive correlation between them. Thus, a linear regression model was build. With the build model it is possible to predict the score that a student will achieve, provided the amount of studied hours per day. The model was trained and the obtained coefficient of correlation is approximately 0.95, which indicates that the regression line fits the observed data points, i.e., that the build model is a good representation of the data.
